```python
def contains_keyword(*args):
    import keyword
    if any([arg for arg in args if keyword.iskeyword(arg)== True]):
        return True
    else:
        return False
```

```python
contains_keyword("home","def")
```

```python
import keyword
 
def contains_keyword(*args):
    for item in args:
        if keyword.iskeyword(item): return True
    return False
```
